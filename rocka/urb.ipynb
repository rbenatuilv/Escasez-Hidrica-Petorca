```python
import geopandas as gpd
import matplotlib.pyplot as plt
import math
```

```python
datos_geo = gpd.read_file("Areas_Pobladas.zip")

# Visualizar los datos en un mapa
datos_geo.plot(figsize=(10, 10), color='blue')
plt.title('Mapa GeoPandas')
plt.xlabel('Longitud')
plt.ylabel('Latitud')
plt.show()
```

```python
datos_geo_comuna_petorca = datos_geo[datos_geo["comuna"] == "Petorca"]
datos_geo_comuna_petorca
```

```python
datos_geo_comuna_petorca.plot(figsize=(10, 10), color='blue')
plt.title('Mapa GeoPandas')
plt.xlabel('Longitud')
plt.ylabel('Latitud')
plt.show()
```

```python
poligono_proyectado = datos_geo_comuna_petorca.to_crs(epsg='3857')
area_poligono_proyectado = poligono_proyectado.geometry.area
list(area_poligono_proyectado)[0]

```

```python
metros_cuadrados = datos_geo_comuna_petorca.area
metros_cuadrados = list(metros_cuadrados)[0]
metros_cuadrados
```

Poblacion de petorca 2017: 9826
Poblacion urbana: 41,8%
Litros de agua por persona 120 a dia 

```python
poblacion_urbana = math.floor(9_826 * 0.418)
poblacion_urbana
```

```python
litros_de_agua_total_urbanano = poblacion_urbana * 120
print(f"{litros_de_agua_total_urbanano:3,} litros de agua")
```

```python
litros_por_m2 = litros_de_agua_total_urbanano / metros_cuadrados
litros_por_m2
```

```python
print(f"{litros_por_m2:.3f} litros de agua por cada m2")
```
